```python
import gradio as gr

# This is the exact output you got from Granite model in the notebook earlier
granite_response = """
1. Problem Statement:
Teenagers in India often lack access to financial education and tools that can help them manage their finances effectively. ...

2. Proposed Solution:
A mobile application called "FinPal" that offers an engaging and interactive platform for teenagers...

3. MVP Features:
- User registration and profile creation
- Interactive financial literacy modules
- Gamified budgeting and saving challenges
- Personalized financial goal setting
- Progress tracking and rewards system
- Integration with parental accounts
- Basic investment education and simulation tools

4. Suggested Tech Stack:
- Frontend: React Native
- Backend: Node.js with Express
- Database: MongoDB
- Cloud: AWS or Google Cloud
- Payments: Razorpay or Paytm
- Analytics: Google Analytics

5. Go-to-Market Strategy:
- Partner with schools and youth organizations
- Use Instagram, TikTok, and YouTube for outreach
- Collaborate with teen influencers
- Conduct webinars and workshops
- Referral programs
- Attend hackathons and student events

6. Monetization Plan:
- Freemium model (basic free, premium for coaching/tools)
- Non-intrusive in-app ads
- Fintech partnerships
- Parental premium accounts
- Merchandise (FinPal-themed items)
"""

# Gradio function that returns the fixed output
def show_blueprint(domain, audience, budget, timeline, country):
    return granite_response

# Create Gradio interface
demo = gr.Interface(
    fn=show_blueprint,
    inputs=[
        gr.Textbox(label="Domain"),
        gr.Textbox(label="Target Audience"),
        gr.Textbox(label="Budget"),
        gr.Textbox(label="Timeline (months)"),
        gr.Textbox(label="Country")
    ],
    outputs="text",
    title="Startup Blueprint Generator (Granite-based)",
    description="Based on IBM Granite output generated via Watsonx Studio"
)

# Launch it inside Watsonx notebook
demo.launch(share=True)  # Shows directly in the notebook
```
